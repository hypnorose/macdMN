```python
import pandas
from matplotlib import pyplot as plt
import numpy as np
from math import *

def rollingAvg(p,N):
    # pierwsze
    divisor = 1
    out = []
    a = (2/(N+1))
    for i in range(N):
        divisor *= (1-a)
        divisor += 1
    for i in range(0,len(p)):
        v = 0
        divisor = 1
        for j in range(-N,0):
            if i+j <= 0:
                continue
            v += p[i+j]
            v *= (1-a)
            divisor *= (1-a)
            divisor += 1
        v += p[i]
        out.append(v/divisor)
            
        
    return out
    
```

```python
data = pandas.read_csv("data.csv")
print(data)
```

```python

plt.rcParams['figure.figsize'] = [16,10]
ema12 = rollingAvg(data['Price'][:400],12)
ema26 = rollingAvg(data['Price'][:400],26)
macd = []
for i in range(len(ema12)):
    macd.append(ema12[i]-ema26[i])
signal = rollingAvg(macd,9)

plt.plot(macd)
plt.plot(signal,color='red')
plt.show()
```
